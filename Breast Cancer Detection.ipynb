## Splitting the dataset train and test set

```python
dataset.head()
```

```python
# matrix of feature/independent variable
x = dataset.iloc[:, 1:-1].values
```

```python
x.shape
```

```python
x.shape
```

```python
y = dataset.iloc[:, -1].values
```

```python
y.shape
```

to split a dataset...we use a class from sklearn library...go to ...sklearn documentation...now use the spitter function

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Building the model

# 1) Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifir_lr = LogisticRegression(random_state=0)
```

```python
classifir_lr.fit(x_train, y_train)
```

```python
y_pred = classifir_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
```

```python
results = pd.DataFrame([['Logistic Regression', acc, f1, prec, rec]],
                      columns = ['Model','Accuracy', 'F1 Score', 'Precision', 'Recall'])
```

```python
results
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

### Cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=classifir_lr, X=x_train, y = y_train, cv=10)
```

```python
print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```
